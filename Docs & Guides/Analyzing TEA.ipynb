## TEA Manipulation tools

### 1. Merging Multiple TEA DataFrames

**Function**: `merge_svo_dataframes(df_list)`

**Purpose**: Merges a list of DataFrames containing TEA data into a single DataFrame. The function ensures unique `svo_id` values across all merged DataFrames.


```python
df_list = [df, df2]

# Merge them
merged_df = wdw_analytics.merge_svo_dataframes(df_list)

# Display the merged DataFrame
display(merged_df.head())
```

### 2. Saving the TEA data to a CSV file.

The TEA data is saved as as a Pandas dataframe, as such, the pandas function `to_csv` can be used to save the data to a CSV file.

### 3. Exporting Elements from the TEA Dataframe


#### Exporting Hypergraphs

**Function**: `export_hypergraphs(df)`

**Purpose**: Extracts syntactic hypergraphs from the DataFrame.



```python
hypergraphs_list = wdw_analytics.export_hypergraphs(df)

print("Number of hypergraphs extracted:", len(hypergraphs_list))
print("Sample hypergraph:", hypergraphs_list[0] if hypergraphs_list else None)
```

#### Exporting Subjects


**Function**: `export_subj(df)`

**Purpose**: Extracts a set of **unique subjects** from the DataFrame. Each element in the set is a tuple of the form `(subject, valence)`, where `valence` can be `'positive'`, `'negative'`, or `'neutral'`, determined by VADER sentiment.


```python
subjects = wdw_analytics.export_subj(df)
print(subjects)
```

#### Exporting Objects

**Function**: `export_obj(df)`

**Purpose**: Extracts a set of **unique objects** from the DataFrame. Each element in the set is a tuple of the form `(object, valence)`.


```python
objects = wdw_analytics.export_obj(df)
print(objects)
```

#### Exporting Verbs


**Function**: `export_verb(df)`

**Purpose**: Extracts a set of **unique verbs** from the DataFrame. Each element in the set is a tuple of the form `(verb, valence)`.


```python
verbs = wdw_analytics.export_verb(df)
print(verbs)
```

### 4. Filtering the SVO DataFrame by TEA

**Function**: `filter_svo_dataframe_by_tea(df, TEA, TEA2=None)`

**Purpose**: Filters the SVO DataFrame to keep rows where `TEA` and `TEA2` match the provided arguments (`"Agent"`, `"Event"`, `"Target"`).  
- If `TEA2` is `None`, it filters by `TEA` alone.  
- Ensures we only look at **syntactic** relationships (`Semantic-Syntactic == 0`).

```python
# Filter to keep only Subject ("Agent") - Verb ("Event") edges
filtered_df_who_did = wdw_analytics.filter_svo_dataframe_by_tea(df, "Agent", "Event")

# Examine the result
display(filtered_df_who_did.head(10))
```

# TEA Analysis tools

Target-Event-Agent (TEA) Networks can be analyzed with standard network science procedures. This library provides convenient functions for filtering, transforming, and analyzing these networks.

## Building Graphs from SVO Data

This library uses NetworkX to build graphs from SVO (Subject-Verb-Object) data. These graphs can then be analyzed with network science. Nodes are labeled with the underlying text (e.g., "gale", "drink", "tasty beer") and typed by their syntactic role ("Agent", "Event", or "Target"). Edges are weighted based on syntactic relations.



**Function**: `svo_to_graph(df, subject_filter=None, object_filter=None)`

**Purpose**  
Convert a DataFrame of SVO data into a NetworkX graph. Optionally filter by specific subjects or objects.  
- Nodes represent subjects, verbs, and objects (with unique IDs like `gale_s` for "gale" and "_s" as a subject).  
- Edges capture relationships:  
  - `relation` is a set of relationship types (`"syntactic"`, `"synonym"`).  
  - `weight` increments for each syntactic occurrence.  
  - `hypergraph` can group edges into higher-level scenarios or sets.

**Parameters**  
- `df` (`pandas.DataFrame`): A TEA Pandas Dataframe.
- `subject_filter` (`function`|`list`|`None`): Filters rows by subject label.  
- `object_filter` (`function`|`list`|`None`): Filters rows by object label.

**Returns**  
- A `networkx.Graph` where each node has:  
  - `label`: Original text (e.g., `"gale"`)  
  - `node_type`: `"subject"`, `"verb"`, or `"object"`  
- Each edge stores:  
  - `relation`: A set of relation types  
  - `hypergraph`: A set of the hypergraphs this node appears in.  
  - `weight`: Integer count of syntactic relations.


```python

G = wdw_analytics.svo_to_graph(df, subject_filter="gale")

# Now you have a networkx Graph G. Nodes and edges contain detailed attributes.
for node in G.nodes(data=True):
    print(node)
for edge in G.edges(data=True):
    print(edge)
```


##  Degree Centrality

In network analysis, degree centrality typically measures the number of edges connected to a node. However, in weighted networks, a node’s weighted degree centrality (also sometimes called “node strength”) is the sum of the weights of its edges. This provides a more nuanced view of a node’s importance than simple degree count.

 

### Weighted Degree Centrality


**Function**: `wdw_weighted_degree_centrality(df, TEA, TEA2=None)`

**Purpose**: 
1. Filters the DataFrame according to `TEA` (and optionally `TEA2`).
2. Builds a graph where edges have a weight for each syntactic relationship.
3. Calculates the normalized node strength (sum of edge weights) as the degree centrality.

```python

# Degree centrality for Subjects ("Agent")
subject_centrality_df = wdw_analytics.wdw_weighted_degree_centrality(df, "Agent")
display(subject_centrality_df)

```

```python
# Degree centrality for Verbs ("Event")
verb_centrality_df = wdw_analytics.wdw_weighted_degree_centrality(df, "Event", None)
display(verb_centrality_df)
```

### Weighted Degree Centrality Overview

**Function**: `wdw_degree_centrality_overview(df)`

**Purpose**: A convenient function that prints an **overview** of the top 20 nodes in each role (Subject, Verb, Object) by their **weighted degree centrality**.


```python
wdw_analytics.wdw_degree_centrality_overview(df)
```

# Putting It All Together

Below is a short example workflow illustrating how you might use these functions in tandem:


```python
import teanets as tea
import teanets.analytics as wdw_analytics

# 1. Extract SVOs from multiple texts
text1 = "Gale loves Karlach. They drink coffee together."
text2 = "Wyll meets Gale. They go to a tavern."
df1 = tea.extract_svos_from_text(text1)
df2 = tea.extract_svos_from_text(text2)

# 2. Merge DataFrames
merged_df = wdw_analytics.merge_svo_dataframes([df1, df2])

# 3. Export unique subjects, objects, and verbs
subjects = wdw_analytics.export_subj(merged_df)
objects = wdw_analytics.export_obj(merged_df)
verbs = wdw_analytics.export_verb(merged_df)

# 4. Filter the merged DataFrame to only subject-verb relationships
df_who_did = wdw_analytics.filter_svo_dataframe_by_tea(merged_df, "Agent", "Event")

# 5. Compute Weighted Degree Centrality for subjects
subject_centrality = wdw_analytics.wdw_weighted_degree_centrality(merged_df, "Agent")
print("Subject Weighted Degree Centrality\n", subject_centrality)

# 6. Get an overview of Weighted Degree Centralities (Subject, Verb, Object)
wdw_analytics.wdw_degree_centrality_overview(merged_df)
```

# Troubleshooting and Tips

1. **Large Texts**: Processing very large texts might be slow due to dependency parsing and coreference resolution. Consider splitting text into smaller chunks.
2. **Coreference Models**: Ensure you have the correct models downloaded if using `stanza` or `fastcoref`. Consider that `stanza` takes up a significant amount of time and computational power compared to the default `fastcoref`.
3. **Inspecting DataFrames**: Always check intermediate DataFrames (e.g., `df.head()`) to ensure expected behavior.
4. **Synonym vs. Syntactic Relationships**:  
   - `Semantic-Syntactic = 1` indicates a semantic (synonym) relationship (e.g., "eggplant" ~ "aubergine").  
   - `Semantic-Syntactic = 0` indicates a syntactic Target-Event-Agent relationship.
